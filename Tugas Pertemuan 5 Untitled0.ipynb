```python
2.#Menampilkan nama sebanyak 200 kali menggunakan for
for i in range(1, 201):
    print(f"{i}.Michael Rafael Nababan")
```

```python
3.#Program menampilkan angka dari 100 sampai 500 menggunakan while

angka = 100
while angka <= 500:
    print(angka)
    angka += 1
```

```python
4.#Menampilkan semua bilangan ganjil dari 1 sampai 200
angka = 1
while angka <= 200:
    if angka % 2 != 0:
        print(angka, end=" ")
    angka += 1
```

```python
5.#Program menampilkan tabel perkalian 9(for)

for i in range(1, 11):
    hasil = 9 * i
    print(f"9 x {i} = {hasil}")
```

```python
# Program menampilkan tabel perkalian 9 (while)

i = 1
while i <= 10:
    hasil = 9 * i
    print(f"9 x {i} = {hasil}")
    i += 1
```
